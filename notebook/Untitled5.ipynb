```python
import pandas as pd
import numpy as np
np.set_printoptions(precision=3,suppress=True)
f=pd.read_csv('2018.csv')
f
```

```python
f.columns=(['a','b','c','d','e','f','g','h','i','j','k','l','m','n','o','p','q','r','s','t','u','v','w','x'])

m=f[f.b=='男']
n=f[f.b=='女']
l=np.array((n.q+n.r)/2)
o=np.array((n.r-n.q)/2)

def p(t,s,x):
    if s==0:
        s=3
    j=1000*(1/s)*np.exp(-((x-t)**2)/(2*s**2))
    return j

w=[]
for (i,j) in zip(l,o):
    for r in np.array(m.f):
        w.append(p(i,j,r))
        
g=np.array(w).reshape(-1,m.index.size)

g[[g<30]]=np.full_like(g[[g<30]],fill_value=30)
g[[g>150]]=np.full_like(g[[g>150]],fill_value=150)

l=np.array((n.s+n.t)/2)
o=np.array((n.t-n.s)/2)
def p2(t,s,x):
    if s==0:
        s=3
    if t>100:
        t=t/2
        s=s/2
    j=1000*(1/s)*np.exp(-((x-t)**2)/(2*s**2))
    return j

w2=[]
for (i,j) in zip(l,o):
    for r in np.array(m.g):
        w2.append(p2(i,j,r))
        
g2=np.array(w2).reshape(-1,m.index.size)

g2[[g2<30]]=np.full_like(g2[[g2<30]],fill_value=30)
g2[[g2>150]]=np.full_like(g2[[g2>150]],fill_value=150)

l=np.array((n.o+n.p)/2)
o=np.array((n.p-n.o)/2)
def p3(t,s,x):
    if s==0:
        s=1
    j=1.3*(20+150*(1/s)*np.exp(-((x-t)**2)/(2*s**2)))
    return j

w5=[]
for (i,j) in zip(l,o):
    for r in np.array(m.c):
        w5.append(p3(i,j,r))
        
g3=np.array(w5).reshape(-1,m.index.size)

g3[[g3<30]]=np.full_like(g3[[g3<30]],fill_value=40)
g3[[g3>150]]=np.full_like(g3[[g3>150]],fill_value=150)

e=n
n=m
m=e
l=np.array((n.q+n.r)/2)
o=np.array((n.r-n.q)/2)
w3=[]
for (i,j) in zip(l,o):
    for r in np.array(m.f):
        w3.append(p(i,j,r))
        
h=np.array(w3).reshape(-1,m.index.size)

h[[h<30]]=np.full_like(h[[h<30]],fill_value=30)
h[[h>150]]=np.full_like(h[[h>150]],fill_value=150)

l=np.array((n.s+n.t)/2)
o=np.array((n.t-n.s)/2)

w4=[]
for (i,j) in zip(l,o):
    for r in np.array(m.g):
        w4.append(p2(i,j,r))
        
h2=np.array(w4).reshape(-1,m.index.size)

h2[[h2<30]]=np.full_like(h2[[h2<30]],fill_value=30)
h2[[h2>150]]=np.full_like(h2[[h2>150]],fill_value=150)

l=np.array((n.o+n.p)/2)
o=np.array((n.p-n.o)/2)

w6=[]
for (i,j) in zip(l,o):
    for r in np.array(m.c):
        w6.append(p3(i,j,r))
        
h3=np.array(w5).reshape(-1,m.index.size)

h3[[h3<30]]=np.full_like(h3[[h3<30]],fill_value=40)
h3[[h3>150]]=np.full_like(h3[[h3>150]],fill_value=150)

df1=pd.DataFrame(g+h.T,index=m.a,columns=n.a)

df2=pd.DataFrame(g2+h2.T,index=m.a,columns=n.a)

df3=pd.DataFrame(g3+h3.T,index=m.a,columns=n.a)

df=df1+df2+df3
c2=[]
c3=[]
for i in np.arange(df.columns.size):
    c1=[]
    for j in np.arange(df.columns.size):
        if df.index.size!=0:
            c1.append(df.iloc[:,j].sort_values()[-1])
        else:
            break
    if c1==[]:
        break
    q=pd.DataFrame(np.array(c1),index=np.array(df.columns))
    ml=df.loc[:,[q.sort_values(by=0,ascending=False).iloc[0].name]]
    b=q.sort_values(by=0,ascending=False).iloc[0,:].name
    c2.append(b)
    c3.append(ml.sort_values(by=b,ascending=False).iloc[0].name)
    if b in df.columns:
        df=df.drop(q.sort_values(by=0,ascending=False).iloc[0,:].name,axis=1)
    if ml.sort_values(by=b,ascending=False).iloc[0].name in df.index:
        df=df.drop(ml.sort_values(by=b,ascending=False).iloc[0].name,axis=0)

df=df1+df2+df3
c4=[]
c5=[]
c6=[]
for (i,j) in zip(c2,c3):
    c4.append(df.get_value(j,i))
    c5.append(f[f.a==i].i.iloc[0])
    c6.append(f[f.a==j].i.iloc[0])

df1=pd.DataFrame(index=['男生','男生QQ','女生','女生QQ','分数'],data=[c2,c5,c3,c6,c4]).T

df.to_csv('123.csv')
df1.to_csv('345.csv')
```

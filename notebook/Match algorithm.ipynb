```python
import pandas as pd
import numpy as np
np.set_printoptions(precision=3,suppress=True)
f=pd.read_csv('2019.csv')
f.iloc[:,1:12]
```

```python
f.columns=(['a','b','c','d','e','f','g','h','i','j','k','l','m','n','o','p','q','r','s','t','u','v','w','x'])
m=f[f.d=='男']
n=f[f.d=='女']
m
```

```python
import random
sjv=list((1+2*random.random())for i in np.arange(n.index.size))
sjan=list((1+2*random.random())for i in np.arange(m.index.size))
```

```python
l=np.array((n.s+n.r)/2)
o=np.array((n.s-n.r)/2)

def p(t,s,x):
    if s<=3:
        s=3
    elif s>=15:
        s=15
    s=(8-6.3)/(15-3)*(s-3)+6.3
    if x<t:
        x=t+(1/1.5)*(x-t)
        j=1000*(1/s)*np.exp(-((x-t)**2)/(2*s**2))
    else:
        j=150-0.6*(150-1000*(1/s)*np.exp(-((x-t)**2)/(2*s**2)))
    return j

w=[]
for (i,j) in zip(l,o):
    for r in np.array(m.h):
        w.append(p(i,j,r))
        
g=np.array(w).reshape(-1,m.index.size)

g[[g<30]]=np.full_like(g[[g<30]],fill_value=30)
g[[g>150]]=np.full_like(g[[g>150]],fill_value=150)

l=np.array((n.u+n.t)/2)
o=np.array((n.u-n.t)/2)
def p2(t,s,x):
    if s<=3:
        s=3
    elif s>=15:
        s=15
    s=(4-3.25)/(15-3)*(s-3)+3.25
    if t>100:
        t=t/2
        s=s/2
    x=t+(1/1.5)*(x-t)
    j=500*(1/s)*np.exp(-((x-t)**2)/(2*s**2))
    return j

w2=[]
for (i,j) in zip(l,o):
    for r in np.array(m.i):
        w2.append(p2(i,j,r))
        
g2=np.array(w2).reshape(-1,m.index.size)

g2[[g2<30]]=np.full_like(g2[[g2<30]],fill_value=30)
g2[[g2>150]]=np.full_like(g2[[g2>150]],fill_value=150)

l=n.q
o=sjan
def p3(t,s,x):
    if s==0:
        s=1
    elif s>=3:
        s=3
    s=(1.7-1.4)/(3-1)*(s-1)+1.4
    x=t+(1/1.5)*(x-t)
    j=1.3*(20+150*(1/s)*np.exp(-((x-t)**2)/(2*s**2)))
    return j

w5=[]
for (i,j) in zip(l,o):
    for r in np.array(m.e):
        w5.append(p3(i,j,r))
        
g3=np.array(w5).reshape(-1,m.index.size)

g3[[g3<30]]=np.full_like(g3[[g3<30]],fill_value=30)
g3[[g3>150]]=np.full_like(g3[[g3>150]],fill_value=150)

e=n
n=m
m=e
l=np.array((n.s+n.r)/2)
o=np.array((n.s-n.r)/2)
w3=[]
for (i,j) in zip(l,o):
    for r in np.array(m.h):
        w3.append(p(i,j,r))
        
h=np.array(w3).reshape(-1,m.index.size)

h[[h<30]]=np.full_like(h[[h<30]],fill_value=30)
h[[h>150]]=np.full_like(h[[h>150]],fill_value=150)

l=np.array((n.u+n.t)/2)
o=np.array((n.u-n.t)/2)

w4=[]
for (i,j) in zip(l,o):
    for r in np.array(m.i):
        w4.append(p2(i,j,r))
        
h2=np.array(w4).reshape(-1,m.index.size)

h2[[h2<30]]=np.full_like(h2[[h2<30]],fill_value=30)
h2[[h2>150]]=np.full_like(h2[[h2>150]],fill_value=150)

l=n.q
o=sjv
w6=[]
for (i,j) in zip(l,o):
    for r in np.array(m.e):
        w6.append(p3(i,j,r))
        
h3=np.array(w5).reshape(-1,m.index.size)

h3[[h3<30]]=np.full_like(h3[[h3<30]],fill_value=30)
h3[[h3>150]]=np.full_like(h3[[h3>150]],fill_value=150)

df1=pd.DataFrame(g+h.T,index=m.c,columns=n.c)

df2=pd.DataFrame(g2+h2.T,index=m.c,columns=n.c)

df3=pd.DataFrame(g3+h3.T,index=m.c,columns=n.c)

df=df1+df2+df3
c2=[]
c3=[]
```

```python
for i in np.arange(df.columns.size):
    c1=[]
    for j in np.arange(df.columns.size):
        if df.index.size!=0:
            c1.append(df.iloc[:,j].sort_values()[-1])
        else:
            break
    if c1==[]:
        break
    q=pd.DataFrame(np.array(c1),index=np.array(df.columns))
    ml=df.loc[:,[q.sort_values(by=0,ascending=False).iloc[0].name]]
    b=q.sort_values(by=0,ascending=False).iloc[0,:].name
    c2.append(b)
    c3.append(ml.sort_values(by=b,ascending=False).iloc[0].name)
    if b in df.columns:
        df=df.drop(q.sort_values(by=0,ascending=False).iloc[0,:].name,axis=1)
    if ml.sort_values(by=b,ascending=False).iloc[0].name in df.index:
        df=df.drop(ml.sort_values(by=b,ascending=False).iloc[0].name,axis=0)

df=df1+df2+df3
c4=[]
c5=[]
c6=[]
for (i,j) in zip(c2,c3):
    c4.append(df.get_value(j,i))
    c5.append(f[f.c==i].k.iloc[0])
    c6.append(f[f.c==j].k.iloc[0])

df1=pd.DataFrame(index=['男生','男生QQ','女生','女生QQ','分数'],data=[c2,c5,c3,c6,c4]).T
df.to_csv('123.csv')
df1.to_csv('345.csv')
```

```python
def find(s):
    if f[f.a==s].size==0:
        return '恭喜你！你已成功匹配到傻子！'
    else:
        if f[f.a==s].iloc[0,1]=='男':
            if df1[df1['男生']==s].size==0:
                return '恭喜你！你已成功匹配到慧慧！'
            else:
                r=df1[df1['男生']==s].iloc[0,2]
                xc=df1[df1['男生']==s].iloc[0,4]
        else:
            if df1[df1['女生']==s].size==0:
                return '恭喜你！你已成功匹配到辉辉！'
            else:
                r=df1[df1['女生']==s].iloc[0,0]
                xc=df1[df1['男生']==s].iloc[0,4]
    return f[f.a==r],xc
```
